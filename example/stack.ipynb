```python
import patchworklib as pw
```

```python
ax_list = []
for i in range(10):
    ax_list.append(pw.Brick(figsize=(2,2), label="ax{}".format(i)))
stacked_axes = pw.stack(ax_list, operator="|", margin=0.2)
stacked_axes.savefig("../stack_example.png")
```

```python
ax10 = pw.Brick(figsize=(2,2), label="ax10")
stacked_axes = stacked_axes["ax0"]/ax10
stacked_axes.savefig()
```

```python
ax11 = pw.Brick(figsize=(2,4), label="ax11")
#ax11.adjust = False
stacked_axes = stacked_axes["ax2"]/ax11
stacked_axes.savefig()
```
